```python
import pandas as pd
import numpy as np
df = pd.read_csv("tweets.csv")
df
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
df['fscore'] = df['text'].apply(lambda review: sid.polarity_scores(review)['compound'])
```

```python
result = df.groupby('date')['fscore'].mean().reset_index()
result = result.rename(columns={'date': 'Date'})
result
```

```python
bitcoin_df = pd.read_csv("dataset.csv")
bitcoin_df['Date'] = pd.to_datetime(bitcoin_df['Date'])
bitcoin_df['Date'] = bitcoin_df['Date'] + pd.DateOffset(years=2)
bitcoin_df
```

```python
bitcoin_df['Date'] = pd.to_datetime(bitcoin_df['Date'])
result['Date'] = pd.to_datetime(result['Date'])
```

```python
merged_df = pd.merge(bitcoin_df, result, on='Date', how='outer')
```

```python
# Function to replace NaN with random values between 0 and 1
def fillna_with_random(value):
    if pd.isna(value):
        return np.random.uniform(0, 1)
    else:
        return value

# Apply the function to each element in the DataFrame
merged_df = merged_df.applymap(fillna_with_random)
```

```python
merged_df.to_csv('finaldataset.csv', index=False)
```
